# Q1. Explain what Tableau is and why it is preferred over traditional spreadsheet tools for data analysis. Also mention any two industries where Tableau is commonly used and explain one use case.
### **Tableau**

**Tableau** is a **data visualization and business intelligence tool** used to analyze data and create interactive dashboards, charts, and reports.

It helps users **convert raw data into visual insights** for better decision-making.

### **Why Tableau is preferred over traditional spreadsheets**

1. **Better Data Visualization**
   Tableau creates interactive charts, dashboards, and maps easily.

2. **Handles Large Data**
   It can analyze millions of rows of data faster than spreadsheets.

3. **Drag and Drop Interface**
   Users can build charts by simply dragging fields.

4. **Interactive Dashboards**
   Multiple charts can be combined into a dashboard for complete analysis.

### **Industries where Tableau is used**
* **Retail Industry** – for analyzing sales and customer trends
* **Healthcare Industry** – for analyzing patient data and hospital performance

### **Example Use Case**

In the **retail industry**, companies use Tableau to analyze **sales by product category, region, and time**.
Managers can identify **which products generate the most profit** and make better marketing decisions.

# Q3. Create a Bar Chart showing Total Sales by Category

### Fields used in the chart

* **Rows:** Category
* **Columns:** Sales
* **Aggregation applied:** SUM (Sum of Sales)

### Steps in Tableau

1. Drag **Category** to the **Rows Shelf**.
2. Drag **Sales** to the **Columns Shelf**.
3. Tableau automatically creates a **Bar Chart**.
4. Sales will be aggregated as **SUM(Sales)** by default.

### Result

The bar chart shows **total sales for each category** such as Furniture, Office Supplies, and Technology.

# Q8. Difference between Worksheet and Dashboard in Tableau?

### Worksheet

A worksheet is a **single visualization or chart** where data analysis is performed.

Example:
Bar chart, line chart, or map.

### Dashboard

A dashboard is a **collection of multiple worksheets displayed together on one screen**.

### Why dashboards are useful for decision-makers

Dashboards help decision-makers by:

* Showing **multiple insights in one place**
* Making data **easy to understand visually**
* Helping managers make **quick decisions**

# Q9. What is Tableau Prep and why is it important

**Tableau Prep** is a tool used for **data cleaning, transformation, and preparation before analysis**.

### Importance in real-world projects

1. Improves **data quality**
2. Removes **errors and duplicates**
3. Prepares data for **accurate analysis**

### Steps to Remove Null Values

1. Identify fields containing null values.
2. Apply a **filter to exclude null values**.
3. Save the cleaned dataset.

### Steps to Remove Duplicate Records

1. Identify duplicate rows.
2. Use the **Remove Duplicates option**.
3. Keep only **unique records**.

### Steps to Change Incorrect Data Types

1. Identify the column with incorrect data type.
2. Select the column.
3. Change the data type (Text, Number, Date etc.).
4. Save the updated dataset.
